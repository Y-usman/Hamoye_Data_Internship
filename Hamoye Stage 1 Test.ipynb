```python
import pandas as pd
file = 'FoodBalanceSheets_E_Africa_NOFLAG.csv'
df = pd.read_csv(file,encoding='latin-1')
df.head()
```

```python
#Question 11
#What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

#Hint: Perform a groupby sum aggregation on ‘Item’

total_sum_of_Animal_Fat_in_2014 = df.groupby('Item')['Y2014'].sum()

total_sum_of_Animal_Fat_in_2017 = df.groupby('Item')['Y2017'].sum()

print(total_sum_of_Animal_Fat_in_2014)
print(total_sum_of_Animal_Fat_in_2017)

# Answer = The total sum of Animal Fat produced in 2014 is 209460.54 and for Year 2017 is 269617.53
```

```python
#Question 12
# What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?
mean_for_y2015 = df['Y2015'].mean()
print(mean_for_y2015)
std_for_y2015 = df['Y2015'].std()
print(std_for_y2015)

# Answer 
# the mean across the whole dataset for the year 2015 to 3 decimal places is 135.236 
# the standard deviation across the whole dataset for the year 2015 to 3 decimal places is 1603.404
```

```python
#Question 13
# What is the total number and percentage of missing data in 2016 to 2 decimal places?

no_of_missing_data = df['Y2016'].isnull().sum()
print(no_of_missing_data)
percentage_of_missing_data = (df['Y2016'].isnull().sum()/ (df['Y2016'].notnull().sum()+df['Y2016'].isnull().sum())) * 100
print(percentage_of_missing_data.)
#Answer = 1535 and 2.52
```
